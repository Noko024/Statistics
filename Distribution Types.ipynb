Distribution Types

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, binom, poisson, expon, uniform, geom

# Function to generate and plot data distributions with explanations
def plot_data_distributions():
    # Parameters for distributions
    x = np.arange(0, 20, 0.1)
    mu, sigma = 10, 2  # Parameters for the normal distribution
    n, p = 20, 0.5     # Parameters for the binomial distribution
    lam = 5            # Parameter for the Poisson distribution
    scale = 1          # Parameter for the exponential distribution
    loc = 5            # Parameters for the uniform distribution
    low = 2
    high = 8
    p_geom = 0.3       # Parameter for the geometric distribution

    # Calculate distributions
    normal_dist = norm.pdf(x, mu, sigma)
    binom_dist = binom.pmf(np.floor(x), n, p)
    poisson_dist = poisson.pmf(np.floor(x), lam)
    exponential_dist = expon.pdf(x, scale=scale)
    uniform_dist = uniform.pdf(x, loc=loc, scale=(high - low))
    geometric_dist = geom.pmf(np.floor(x), p_geom)

    # Create a figure with a dark background
    plt.figure(figsize=(24, 12), facecolor='#05151c')

    # Plot the Normal distribution
    plt.subplot(2, 3, 1, facecolor='#05151c')
    plt.plot(x, normal_dist, color='orange')
    plt.scatter([mu], [norm.pdf(mu, mu, sigma)], color='orange')
    for j in range(10):
        plt.plot(x, normal_dist - (j + 1) * 0.01, color='gray', alpha=0.2)
        plt.plot(x, normal_dist + (j + 1) * 0.01, color='gray', alpha=0.2)
    plt.gca().spines['top'].set_color('none')
    plt.gca().spines['right'].set_color('none')
    plt.gca().spines['bottom'].set_color('gray')
    plt.gca().spines['left'].set_color('gray')
    plt.gca().tick_params(axis='x', colors='gray')
    plt.gca().tick_params(axis='y', colors='gray')
    plt.title('Normal Distribution', color='white')
    plt.text(10, 0.15, 'A continuous probability distribution \nthat is symmetric about the mean, \nshowing that data near the mean \nare more frequent in occurrence.', 
             color='white', bbox=dict(facecolor='gray', alpha=0.2))

    # Plot the Binomial distribution
    plt.subplot(2, 3, 2, facecolor='#05151c')
    plt.plot(x, binom_dist, color='cyan')
    plt.scatter([n * p], [binom.pmf(n * p, n, p)], color='cyan')
    for j in range(10):
        plt.plot(x, binom_dist - (j + 1) * 0.01, color='gray', alpha=0.2)
        plt.plot(x, binom_dist + (j + 1) * 0.01, color='gray', alpha=0.2)
    plt.gca().spines['top'].set_color('none')
    plt.gca().spines['right'].set_color('none')
    plt.gca().spines['bottom'].set_color('gray')
    plt.gca().spines['left'].set_color('gray')
    plt.gca().tick_params(axis='x', colors='gray')
    plt.gca().tick_params(axis='y', colors='gray')
    plt.title('Binomial Distribution', color='white')
    plt.text(10, 0.2, 'A discrete probability distribution \nthat summarizes the number of successes \nin a fixed number of trials, \neach with the same probability of success.', 
             color='white', bbox=dict(facecolor='gray', alpha=0.2))

    # Plot the Poisson distribution
    plt.subplot(2, 3, 3, facecolor='#05151c')
    plt.plot(x, poisson_dist, color='red')
    plt.scatter([lam], [poisson.pmf(lam, lam)], color='red')
    for j in range(10):
        plt.plot(x, poisson_dist - (j + 1) * 0.01, color='gray', alpha=0.2)
        plt.plot(x, poisson_dist + (j + 1) * 0.01, color='gray', alpha=0.2)
    plt.gca().spines['top'].set_color('none')
    plt.gca().spines['right'].set_color('none')
    plt.gca().spines['bottom'].set_color('gray')
    plt.gca().spines['left'].set_color('gray')
    plt.gca().tick_params(axis='x', colors='gray')
    plt.gca().tick_params(axis='y', colors='gray')
    plt.title('Poisson Distribution', color='white')
    plt.text(5, 0.2, 'A discrete probability distribution \nexpressing the probability of a given number \nof events occurring in a fixed interval of time \nor space if these events happen \nwith a known constant mean rate.', 
             color='white', bbox=dict(facecolor='gray', alpha=0.2))

    # Plot the Exponential distribution
    plt.subplot(2, 3, 4, facecolor='#05151c')
    plt.plot(x, exponential_dist, color='green')
    plt.scatter([scale], [expon.pdf(scale)], color='green')
    for j in range(10):
        plt.plot(x, exponential_dist - (j + 1) * 0.01, color='gray', alpha=0.2)
        plt.plot(x, exponential_dist + (j + 1) * 0.01, color='gray', alpha=0.2)
    plt.gca().spines['top'].set_color('none')
    plt.gca().spines['right'].set_color('none')
    plt.gca().spines['bottom'].set_color('gray')
    plt.gca().spines['left'].set_color('gray')
    plt.gca().tick_params(axis='x', colors='gray')
    plt.gca().tick_params(axis='y', colors='gray')
    plt.title('Exponential Distribution', color='white')
    plt.text(5, 0.6, 'A continuous probability distribution \nthat describes the time between events \nin a Poisson process, \nwhere events occur continuously \nand independently at a constant average rate.',
             color='white', bbox=dict(facecolor='gray', alpha=0.2))

    # Plot the Uniform distribution
    plt.subplot(2, 3, 5, facecolor='#05151c')
    plt.plot(x, uniform_dist, color='blue')
    plt.scatter([loc + (high - low) / 2], [uniform.pdf(loc + (high - low) / 2, loc=loc, scale=(high - low))], color='blue')
    for j in range(10):
        plt.plot(x, uniform_dist - (j + 1) * 0.01, color='gray', alpha=0.2)
        plt.plot(x, uniform_dist + (j + 1) * 0.01, color='gray', alpha=0.2)
    plt.gca().spines['top'].set_color('none')
    plt.gca().spines['right'].set_color('none')
    plt.gca().spines['bottom'].set_color('gray')
    plt.gca().spines['left'].set_color('gray')
    plt.gca().tick_params(axis='x', colors='gray')
    plt.gca().tick_params(axis='y', colors='gray')
    plt.title('Uniform Distribution', color='white')
    plt.text(5, 0.2, 'A continuous probability distribution \nwhere each value within a certain range \nis equally likely to occur.', 
             color='white', bbox=dict(facecolor='gray', alpha=0.2))

    # Plot the Geometric distribution
    plt.subplot(2, 3, 6, facecolor='#05151c')
    plt.plot(x, geometric_dist, color='purple')
    plt.scatter([0], [geom.pmf(0, p_geom)], color='purple')
    for j in range(10):
        plt.plot(x, geometric_dist - (j + 1) * 0.01, color='gray', alpha=0.2)
        plt.plot(x, geometric_dist + (j + 1) * 0.01, color='gray', alpha=0.2)
    plt.gca().spines['top'].set_color('none')
    plt.gca().spines['right'].set_color('none')
    plt.gca().spines['bottom'].set_color('gray')
    plt.gca().spines['left'].set_color('gray')
    plt.gca().tick_params(axis='x', colors='gray')
    plt.gca().tick_params(axis='y', colors='gray')
    plt.title('Geometric Distribution', color='white')
    plt.text(5, 0.3, 'A discrete probability distribution \nthat models the number of trials needed \nuntil the first success occurs \nin a series of independent Bernoulli trials, \nwith a constant probability of success.', 
             color='white', bbox=dict(facecolor='gray', alpha=0.2))

    # Adjust layout
    plt.tight_layout()

    # Save and show the plot
    plt.savefig('distribution_plot_with_explanations.png', facecolor='#05151c')
    plt.show()

# Call the function to plot data distributions with explanations
plot_data_distributions()
```
